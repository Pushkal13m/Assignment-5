```python
import glob
import numpy 
import matplotlib
import pandas as pd
import matplotlib.pyplot 
import matplotlib.pyplot as plt
import numpy as np
dataset = sorted(glob.glob('F:\Assignment#5\data*.csv'))
print(dataset)
```

```python
def visualize(filename):
    data = pd.read_csv(filepath_or_buffer = filename, header=None)
    df=data.describe()
    df.loc['mean'].plot()
    df.loc['min'].plot()
    df.loc['max'].plot()
    plt.legend(["mean","min","max"], loc ="upper right")
    plt.xlabel("features")
    plt.ylabel("Values")
    matplotlib.pyplot.show()
```

```python
def identify_issues(filename):
    data = pd.read_csv(filepath_or_buffer = filename, header=None)
    if data.iloc[:, 0].max() == 0 and data.iloc[:, 20].max() == 20 :
        print("I doubt the integrity of this dataset") 
    elif sum(data.min()) == 0:
        print("Why is this happening?")
    else:
        print("Dataset looks good")
```

```python
for datasets in dataset[:4]:
    print("\n")
    print(datasets)
    visualize(datasets)
    identify_issues(datasets)
```
